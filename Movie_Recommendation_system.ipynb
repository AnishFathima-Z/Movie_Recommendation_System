Importing the Dependencies

```python
import numpy as np
import pandas as pd
import difflib    #get the closest match of the movie given by the user
from sklearn.feature_extraction.text import TfidfVectorizer#converting the textual data to numerical format
from sklearn.metrics.pairwise import cosine_similarity #to identify simlarity score
```

Data Collection and Preprocessing

```python
#Loading the data from CSV file to the pandas data file
movies_data =pd.read_csv('/content/movies.csv')
```

```python
#Printing the 5 rows of the dataframe
movies_data.head()
```

```python
# to find the no.of.rows and columns in the dataframe
movies_data.shape
```

```python
#Selecting the relavent features for the recommendation
selected_features =['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#Replacing the null values using null string

for feature in selected_features:
  movies_data[feature]=movies_data[feature].fillna('')
```

```python
# combining all the 5 selected_feature taken above
combined_features = movies_data['genres'] + ' ' + movies_data['keywords'] + ' ' + movies_data['tagline'] + ' ' + movies_data['cast'] + ' ' + movies_data['director']
```

```python
print(combined_features)
```

```python
# Converting text data to feature vector
vectorizer =TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity


```python
# Getting the similarity score using cosine similarity
similarity =cosine_similarity(feature_vectors)# to find the related movies scores
print(similarity)
```

```python
print(similarity.shape)# values are how much the movie is similar to other movie
```

```python
#Getting the movie name from the user
movie_name=input("Enter your favourite movie name:")
```

```python
#Creating the list with all the movies names given in the dataset

list_of_all_titles=movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#Finding the close match of the movie name given by the user

find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
#finding the first close match only one
close_match=find_close_match[0]
print(close_match)
```

```python
#Find the index of the movie with title(to find the similarity score later)
index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#Getting the list of similar movies
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
# Higher similarity score is needed so,sort from higher to lower based on similarity score

sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)# key=lambda x means similarity score and x[1]means similarity score(0.13856182473796614) and prinitng them in desending order so ,reverse=True
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index
print("Movies suggested for you:\n")
i=1
for movie in sorted_similar_movies:
  index=movie[0]#movie[0] represents 326 from above o/p
  title_from_index=movies_data[movies_data.index==index]['title'].values[0]
  if i<30:#recommendation of the movie related
    print(i,'.',title_from_index)
    i+=1
```

Movie Recommendation System
